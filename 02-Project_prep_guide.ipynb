## AI Usecase 

First, you need to find a business case worth solving! Try to spend some time looking for a problem where AI could help. 

If you're having trouble, narrow it down to what AI is good at. Usually, people use AI for **optimization** or **automation**. 

For example: 

* *You might be a retail company willing to optimize its promotion process* ==> Starbucks works on this case and is using AI to predict which product to apply a discount on. 

* *Detect churn* ==> Widely spread use case that a lot of companies are trying to solve (From IBM to Sanofi)

Finally, think about the technologies that you would need to implement to create your AI Product. Python is not the only answer. You might use Cloud Platforms, servers, deployment technologies... During your bootcamp, you are free to use any type of technology you want, but you might have constraints in your job.

### MVP Design 

MVP stands for Minimum Viable Product and it comes again from the [Lean Startup](https://en.wikipedia.org/wiki/Lean_startup) from Eric Ries. 

From the author's word, an MVP is : 

*A version of a new product which allows a team to collect the maximum amount of validated learning about customers with the least effort*

In other words, the idea of an MVP is to build a product that includes all the learnings that you collected in a minimum amount of time. 

The whole balance is to make something working without spending too much time on it because you want to put it in your end-users' hands as fast as possible to learn from them and iterate on the product. 

As a person working on an AI project, your MVP should: 

* Give you acceptable predictions
* These predictions should be useful 
* End-users should be able to use it in a way or another

🚨🚨 But again, don't spend too much time on it. The whole idea is actually to make something working not something perfect! 🚨🚨


### Start small grow big ! 🌴🌴

It's extremely important that you reduce your project's scope! Otherwise you won't be able to make anything work. 

Let's say for example that you want to build an AI that detects sign language. You should focus on a few signs first before expanding the scope to the whole language. 

Same thing for your architecture. Don't go right off the bat with Kubernetes, Spark and hardcore Pytorch! Instead try something locally then stage one application after the other. 

Another way of putting is **think microservices**. The more you can break things down into independant services, the easier you architecture will be to manage.

### Collect Data & Build a model 

Here are the steps that you've been working on for quite a bit of time now. What is important to respect is the iterative process between collecting data and building a model. 

<img src="https://full-stack-assets.s3.eu-west-3.amazonaws.com/images/Collect_Data_and_model.png" alt="Data Collection & Modeling" width="60%" />

The first thing you need to do is to collect data. You should have defined in the MVP Design step, the data you need to build your model. What we advise is for you to **collect as little data as possible**. The reason why is because data is hard to collect, you don't want to spend all your time on it. Once you have a reasonable dataset, you should work on it and start building a model.  

Let's be blunt, you won't have a great model at first. Most likely it will overfit or underfit. Don't panic, it's completely normal. What it means is that you need to iterate over your model and your data. 

First, try to go back to the data collection step and **get more observations**. Then do more exploration (EDA), you need to understand your dataset better before re-applying a model on it. 

Then apply your model and fine tune it. You should already get better results.

You now went through the whole iteration loop. You can now repeat the whole process until you get acceptable results. 

🚨🚨 During your bootcamp as well as in your company, you shouldn't get stuck in this iteration process. You should set a maximum number of loops you are going to go through, stop once you hit that number and start thinking about the value that your project is bringing on the table. 

*   If are achieving acceptable results, you should go to the next step 
*   If you can't achieve acceptable results, it's okay and normal. You should analyze as to why, produce recommandations and tackle a new project. 

### Build an API 

APIs are extremely useful to access data. And, it includes AI models. Imagine simply calling a url, attach some input data to it and get instant prediction. Pretty cool right? 

It's also much easier to manage afterwards. Our advice is to use MLFlow directly but you can also build a Flask or FastAPI application depending on your needs. 


### Docker Docker Docker

Everything is simpler with Docker. It doesn't seem like it at the beginning but trust me, when dealing with production, it is your best ally. The sooner you dockerize everything, the better.
